```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn import metrics
Spe1 = [0, 0.917, 0.917, 0.917, 0.958, 0.958, 0.958, 0.958, 0.958, 1, 1]
FPR1 = [1, 0.083, 0.083, 0.083, 0.042, 0.042, 0.042, 0.042, 0.042, 0, 0]
TPR1 = [1, 0.837, 0.818, 0.810, 0.810, 0.810, 0.810, 0.810, 0.783, 0.756, 0]

#auc1 = metrics.auc(TPR1, FPR1)
auc1 = -1 * np.trapz(TPR1, FPR1)

plt.plot(FPR1, TPR1, linestyle='-', color='darkred', lw = 2, label='ROC for fold 1', clip_on=False)
plt.plot([0, 1], [0, 1], color='navy', linestyle='--', label='Random / No Skill')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate (1-Specificity)', fontsize=13)
plt.ylabel('True Positive Rate (Sensitivity)', fontsize=13)
plt.title('ROC curve, AUC = %.2f'%auc1, fontsize=16)
plt.legend(loc="lower right")
plt.savefig('AUC fold 1.png')
plt.show()
```
